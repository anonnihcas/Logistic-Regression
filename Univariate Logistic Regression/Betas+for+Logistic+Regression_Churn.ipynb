```python
import pandas as pd
import numpy as np
```

```python
# dib = pd.read_csv('Diabetes+Example+Data.csv')
dib = pd.read_csv('churn_data.csv')
```

```python
dib.head()
```

```python
# Converting Yes to 1 and No to 0
dib['Churn'] = dib['Churn'].map({'Yes': 0, 'No': 1})
```

```python
dib.head()
```

```python
variables_in_x = ['customerID','tenure','PhoneService','Contract','PaperlessBilling','PaymentMethod','MonthlyCharges','TotalCharges']
# Putting feature variable to X
X = dib[variables_in_x]

# Putting response variable to y
y = dib['Churn']
```

```python

# Creating a dummy variable for some of the categorical variables and dropping the first one.
dummy1 = pd.get_dummies(dib[['Contract', 'PaymentMethod']], drop_first=True)

# Adding the results to the master dataframe
dib = pd.concat([dib, dummy1], axis=1)


# List of variables to map

# varlist =  ['PhoneService', 'PaperlessBilling','PaymentMethod_Credit card (automatic)','PaymentMethod_Electronic check','PaymentMethod_Mailed check','Contract_One year','Contract_Two year']
varlist =  ['PhoneService', 'PaperlessBilling']


# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the housing list
dib[varlist] = dib[varlist].apply(binary_map)

varlist =  ['PaymentMethod_Credit card(automatic)','PaymentMethod_Electroniccheck','PaymentMethod_Mailedcheck','Contract_Oneyear','Contract_Twoyear']

# Defining the map function
def convert_bool_to_int(x):
    return x.map({False:0, True: 1})

dib[varlist].apply(convert_bool_to_int)

dib.drop(columns=['PaymentMethod','Contract','customerID'])
# dib.drop('Contract')
# dib.drop('customerID')
```
